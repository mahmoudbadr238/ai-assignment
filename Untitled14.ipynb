```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.set_printoptions(suppress=True, linewidth=120)

USE_UPLOADER = True
if USE_UPLOADER:
    from google.colab import files
    uploaded = files.upload()
    csv_path = list(uploaded.keys())[0]
else:
    csv_path = "your.csv"

df = pd.read_csv(csv_path)
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
assert len(numeric_cols) >= 2, "Need at least 2 numeric columns: features + target."
target_col = numeric_cols[-1]
feature_cols = numeric_cols[:-1]
X = df[feature_cols].values.astype(float)
y = df[target_col].values.astype(float)

def standardize_fit(X):
    mu = X.mean(axis=0)
    sigma = X.std(axis=0, ddof=0)
    sigma = np.where(sigma == 0, 1.0, sigma)
    return mu, sigma

def standardize_transform(X, mu, sigma):
    return (X - mu) / sigma

mu, sigma = standardize_fit(X)
Xs = standardize_transform(X, mu, sigma)

def split_idx(n, size, seed):
    idx = np.arange(n)
    rng = np.random.default_rng(seed)
    rng.shuffle(idx)
    cut = int(np.ceil(size * n))
    return idx[:cut], idx[cut:]

idx_test, idx_trainfull = split_idx(Xs.shape[0], 0.2, 42)
X_test, y_test = Xs[idx_test], y[idx_test]
X_trainfull, y_trainfull = Xs[idx_trainfull], y[idx_trainfull]
idx_val, idx_train = split_idx(X_trainfull.shape[0], 0.15, 123)
X_val, y_val = X_trainfull[idx_val], y_trainfull[idx_val]
X_train, y_train = X_trainfull[idx_train], y_trainfull[idx_train]

class SGDLinearRegression:
    def __init__(self, lr=0.01, epochs=1000, batch_size=1, shuffle=True, seed=0, verbose=True,
                 early_stopping=True, patience=6, min_delta=1e-6):
        self.lr = lr
        self.epochs = epochs
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.seed = seed
        self.verbose = verbose
        self.early_stopping = early_stopping
        self.patience = patience
        self.min_delta = min_delta
        self.w = None
        self.b = 0.0
        self.loss_history = []
        self.val_history = []
        self.best_epoch = None
        self.stopped_early = False

    @staticmethod
    def mse(y_true, y_pred):
        return float(np.mean((y_true - y_pred) ** 2))

    def predict(self, X):
        return X @ self.w + self.b

    def fit(self, X, y, X_val=None, y_val=None):
        n, d = X.shape
        self.w = np.zeros(d, dtype=float)
        self.b = 0.0
        rng = np.random.default_rng(self.seed)
        best_val = np.inf
        best_w = None
        best_b = None
        patience_left = self.patience

        for epoch in range(1, self.epochs + 1):
            if self.shuffle:
                perm = rng.permutation(n)
                X = X[perm]
                y = y[perm]

            batch_losses = []
            for start in range(0, n, self.batch_size):
                end = min(start + self.batch_size, n)
                Xb = X[start:end]
                yb = y[start:end]
                preds = Xb @ self.w + self.b
                errors = preds - yb
                grad_w = (Xb.T @ errors) / Xb.shape[0]
                grad_b = errors.mean()
                self.w -= self.lr * grad_w
                self.b -= self.lr * grad_b
                batch_losses.append((errors ** 2).mean())

            train_loss = float(np.mean(batch_losses))
            self.loss_history.append(train_loss)

            if X_val is not None and y_val is not None:
                val_loss = self.mse(y_val, self.predict(X_val))
                self.val_history.append(val_loss)
                improved = (best_val - val_loss) > self.min_delta
                if improved:
                    best_val = val_loss
                    best_w = self.w.copy()
                    best_b = float(self.b)
                    self.best_epoch = epoch
                    patience_left = self.patience
                else:
                    patience_left -= 1
                    if patience_left <= 0:
                        self.stopped_early = True
                        if best_w is not None:
                            self.w = best_w
                            self.b = best_b
                        break

            if self.verbose and epoch % (self.epochs // 10) == 0:
                print(f"Epoch {epoch} | Train MSE: {train_loss:.6f} | Val MSE: {self.val_history[-1]:.6f}")

        return self

lr = 0.01
epochs = 1000
batch_size = 1
model = SGDLinearRegression(lr=lr, epochs=epochs, batch_size=batch_size, shuffle=True,
                            seed=123, verbose=True, early_stopping=True, patience=6, min_delta=1e-6)
model.fit(X_train.copy(), y_train.copy(), X_val=X_val, y_val=y_val)

train_pred = model.predict(X_train)
val_pred = model.predict(X_val)
test_pred = model.predict(X_test)

train_mse = np.mean((train_pred - y_train) ** 2)
val_mse = np.mean((val_pred - y_val) ** 2)
test_mse = np.mean((test_pred - y_test) ** 2)

print("\nTrain MSE:", train_mse)
print("Val MSE:", val_mse)
print("Test MSE:", test_mse)

plt.figure(figsize=(6,4))
plt.plot(model.loss_history, label="Train MSE")
plt.plot(model.val_history, label="Val MSE")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.title("MSE vs Epochs (Early Stopping)")
plt.grid(True)
plt.legend()
plt.show()
```
